## Download Historic Data
**Note:** With Nasdaq listed companies and other exchange listed companies we will need to separate the ETFs from the individual stocks later on. To make the script that does the separation work later, you will need to do the following:      
1. Run the script to download nasdaq listed companies
2. Run the script to validate data of nasdaq listed companies
3. Run the script for separating the stocks and etfs
4. Then do the same thing but for other exchange listed companies

### Nasdaq Listed Companies

```python
%%time

limit = limit if limit else len(nasdaq_symbols)
end = min(offset + limit, len(nasdaq_symbols))
nasdaq_is_valid = [False] * len(nasdaq_symbols)
# force silencing of verbose API
with open(os.devnull, 'w') as devnull:
    with contextlib.redirect_stdout(devnull):
        for i in range(offset, end):
            nasdaq_s = nasdaq_symbols[i]
            nasdaq_data = yf.download(nasdaq_s, period=period)
            if len(nasdaq_data.index) == 0:
                continue
        
            nasdaq_is_valid[i] = True
            nasdaq_data.to_csv('/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/{}.csv'.format(nasdaq_s))

print('Total number of valid symbols downloaded = {}'.format(sum(nasdaq_is_valid)))
```

### Other Exchange Listed Companies

```python
%%time

limit = limit if limit else len(other_symbols)
end = min(offset + limit, len(other_symbols))
other_is_valid = [False] * len(other_symbols)
# force silencing of verbose API
with open(os.devnull, 'w') as devnull:
    with contextlib.redirect_stdout(devnull):
        for j in range(offset, end):
            other_s = other_symbols[j]
            other_data = yf.download(other_s, period=period)
            if len(other_data.index) == 0:
                continue
        
            other_is_valid[j] = True
            other_data.to_csv('/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/{}.csv'.format(other_s))

print('Total number of valid symbols downloaded = {}'.format(sum(other_is_valid)))
```

## Separating ETFs and Stocks

```python
nasdaq_etfs = nasdaq_valid_data[nasdaq_valid_data['ETF'] == 'Y']['Symbol'].tolist()
nasdaq_stocks = nasdaq_valid_data[nasdaq_valid_data['ETF'] == 'N']['Symbol'].tolist()
```

```python
other_etfs = other_valid_data[other_valid_data['ETF'] == 'Y']['ACT Symbol'].tolist()
other_stocks = other_valid_data[other_valid_data['ETF'] == 'N']['ACT Symbol'].tolist()
```

```python
def nasdaq_move_symbols(nasdaq_symbols, dest):
    for nasdaq_s in nasdaq_symbols:
        nasdaq_filename = '{}.csv'.format(nasdaq_s)
        shutil.move(join('/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/', nasdaq_filename), join(dest, nasdaq_filename))
        
nasdaq_move_symbols(nasdaq_etfs, "/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/etfs/")
nasdaq_move_symbols(nasdaq_stocks, "/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/stocks/")
```

```python
def other_move_symbols(other_symbols, dest):
    for other_s in other_symbols:
        other_filename = '{}.csv'.format(other_s)
        shutil.move(join('/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/', other_filename), join(dest, other_filename))
        
other_move_symbols(other_etfs, "/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/etfs/")
other_move_symbols(other_stocks, "/home/eadric_ricmund/Documents/git_repos/Market_AI/Data/hist/stocks/")
```
